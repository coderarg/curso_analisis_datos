```python
import pandas as pd
```

```python
# Series de pandas

numeros = [3,4,5,6,7]
serie = pd.Series(numeros)
serie, type(serie)
```

```python
# Crear un diccionario

data = {
    "Nombre": ["Juan", "Ana", "Pedro", "Cristian", "Lucia"],
    "Edad": [28, 24, 35, 30, 22],
    "Ciudad": ["Madrid", "Barcelona", "Valencia", "Sevilla", "Bilbao"]
}

data, type(data)
```

```python
# Generar un DataFrame a partir de un diccionario

df = pd.DataFrame(data)
df
```

```python
# Exportar dataframe a csv

df.to_csv("mis_datos.csv")
```

```python
# Importar datos desde un archivo csv

mis_datos = pd.read_csv("mis_datos.csv", index_col=0)
mis_datos
```

```python
# Seleccionar una columna

nombres = mis_datos["Nombre"]
nombres
```

```python
# Seleccionar varias columnas

mis_datos[["Nombre", "Ciudad"]]
```

```python
# Seleccionar una fila

fila = mis_datos.loc[2]
fila
```

```python
# Filtrar por una condición
mis_datos[mis_datos["Edad"]>23]
```

```python
filtro = (mis_datos["Edad"] > 23) & (mis_datos["Nombre"].str.startswith("A"))
mis_datos[filtro]
```

```python
# Filtrar personas mayores de 23 años por query

mis_datos.query("Edad > 23")
mis_datos[mis_datos["Nombre"].isin(["Ana", "Juan"])]

```

```python
def nombres_long_5(nombre):
    return len(nombre) == 5

mis_datos[mis_datos["Nombre"].apply(nombres_long_5)]
```

```python
# Filtrar por rango

mis_datos[mis_datos["Edad"].between(25,35)]
```
