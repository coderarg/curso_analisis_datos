```python
import pandas as pd
```

```python
# Series de pandas

numeros = [3,4,5,6,7]
serie = pd.Series(numeros)
serie, type(serie)
```

```python
# Crear un diccionario

data = {
    "Nombre": ["Juan", "Ana", "Pedro", "Cristian", "Lucia"],
    "Edad": [28, 24, 35, 30, 22],
    "Ciudad": ["Madrid", "Barcelona", "Valencia", "Sevilla", "Bilbao"]
}

data, type(data)
```

```python
# Generar un DataFrame a partir de un diccionario

df = pd.DataFrame(data)
df
```

```python
# Exportar dataframe a csv

df.to_csv("mis_datos.csv")
```

```python
# Importar datos desde un archivo csv

mis_datos = pd.read_csv("mis_datos.csv", index_col=0)
mis_datos
```
